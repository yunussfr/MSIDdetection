```python
# ============================================================================
# Vision Transformer (ViT) - Google Colab
# ============================================================================

# 1. KÜTÜPHANE KURULUMU VE İMPORT
# ============================================================================
!pip install -q transformers datasets pillow scikit-learn matplotlib seaborn

import os
import shutil
import random
from pathlib import Path
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm.auto import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

from transformers import ViTForImageClassification, ViTImageProcessor
from torch.optim import AdamW # AdamW'yi buradan içe aktarın
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# GPU kontrolü
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"🚀 Kullanılan cihaz: {device}")
if torch.cuda.is_available():
    print(f"   GPU: {torch.cuda.get_device_name(0)}")

# Seed ayarı (tekrarlanabilirlik için)
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)


# 2. VERİ SETİ HAZIRLAMA VE BÖLÜTLEME
# ============================================================================

def prepare_dataset(source_dir='/content/drive/MyDrive/Monkeypox Skin Image Dataset',
                   target_dir='/content/split_dataset/',
                   train_ratio=0.70,
                   val_ratio=0.15,
                   test_ratio=0.15):
    """
    Kaynak dizindeki sınıf klasörlerini tarayıp stratified split yapar.
    """

    source_path = Path(source_dir)
    target_path = Path(target_dir)

    # Hedef klasörleri temizle ve oluştur
    if target_path.exists():
        shutil.rmtree(target_path)

    for split in ['train', 'val', 'test']:
        (target_path / split).mkdir(parents=True, exist_ok=True)

    # Sınıfları otomatik tespit et
    classes = [d.name for d in source_path.iterdir() if d.is_dir()]
    classes.sort()

    print(f"📂 Tespit edilen sınıflar: {classes}")
    print(f"📊 Bölütleme oranları - Train: {train_ratio*100}%, Val: {val_ratio*100}%, Test: {test_ratio*100}%\n")

    all_files = []
    all_labels = []

    # Tüm dosyaları topla
    for class_idx, class_name in enumerate(classes):
        class_path = source_path / class_name
        images = list(class_path.glob('*.*'))
        images = [img for img in images if img.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']]

        all_files.extend(images)
        all_labels.extend([class_name] * len(images))

        print(f"   {class_name}: {len(images)} görüntü")

    print(f"\n✅ Toplam {len(all_files)} görüntü bulundu\n")

    # Stratified split
    # Önce train ve temp (val+test) olarak ayır
    train_files, temp_files, train_labels, temp_labels = train_test_split(
        all_files, all_labels,
        test_size=(val_ratio + test_ratio),
        stratify=all_labels,
        random_state=42
    )

    # Temp'i val ve test olarak ayır
    val_ratio_adjusted = val_ratio / (val_ratio + test_ratio)
    val_files, test_files, val_labels, test_labels = train_test_split(
        temp_files, temp_labels,
        test_size=(1 - val_ratio_adjusted),
        stratify=temp_labels,
        random_state=42
    )

    # Dosyaları kopyala
    def copy_files(file_list, label_list, split_name):
        for file_path, label in tqdm(zip(file_list, label_list),
                                     total=len(file_list),
                                     desc=f"📁 {split_name.upper()} kopyalanıyor"):
            dest_dir = target_path / split_name / label
            dest_dir.mkdir(parents=True, exist_ok=True)
            shutil.copy2(file_path, dest_dir / file_path.name)

    copy_files(train_files, train_labels, 'train')
    copy_files(val_files, val_labels, 'val')
    copy_files(test_files, test_labels, 'test')

    # İstatistikleri göster
    print("\n" + "="*60)
    print("📊 BÖLÜTLEME İSTATİSTİKLERİ")
    print("="*60)
    for split_name, labels in [('Train', train_labels),
                                ('Val', val_labels),
                                ('Test', test_labels)]:
        print(f"\n{split_name}:")
        counter = Counter(labels)
        for class_name in classes:
            print(f"   {class_name}: {counter[class_name]} görüntü")
        print(f"   TOPLAM: {len(labels)}")

    return classes, str(target_path)


# Veri setini hazırla
class_names, dataset_path = prepare_dataset()
num_classes = len(class_names)


# 3. CUSTOM DATASET SINIFI
# ============================================================================

class IrisDataset(Dataset):
    def __init__(self, root_dir, split='train', processor=None):
        self.root_dir = Path(root_dir) / split
        self.processor = processor
        self.class_names = sorted([d.name for d in self.root_dir.iterdir() if d.is_dir()])
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.class_names)}

        # Tüm görüntü yollarını ve etiketlerini topla
        self.samples = []
        for class_name in self.class_names:
            class_path = self.root_dir / class_name
            images = list(class_path.glob('*.*'))
            images = [img for img in images if img.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']]

            for img_path in images:
                self.samples.append((str(img_path), self.class_to_idx[class_name]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        image = Image.open(img_path).convert('RGB')

        if self.processor:
            # ViT processor kullan
            encoding = self.processor(images=image, return_tensors="pt")
            pixel_values = encoding['pixel_values'].squeeze()
        else:
            # Manuel transform
            pixel_values = transforms.ToTensor()(image)

        return {
            'pixel_values': pixel_values,
            'labels': torch.tensor(label, dtype=torch.long)
        }


# 4. MODEL VE PROCESSOR HAZIRLAMA
# ============================================================================

print("\n" + "="*60)
print("🤖 MODEL YÜKLEME")
print("="*60)

model_name = "google/vit-base-patch16-224-in21k"
print(f"📥 Model indiriliyor: {model_name}")

# Image processor (224x224 resize ve normalize otomatik)
processor = ViTImageProcessor.from_pretrained(model_name)

# Model yükle
model = ViTForImageClassification.from_pretrained(
    model_name,
    num_labels=num_classes,
    ignore_mismatched_sizes=True  # Son katman değiştiği için
)

# Modeli GPU'ya taşı
model = model.to(device)

print(f"✅ Model başarıyla yüklendi ve {device} cihazına taşındı")
print(f"🎯 Sınıf sayısı: {num_classes}")
print(f"📐 Girdi boyutu: 224x224")


# 5. DATASET VE DATALOADER OLUŞTURMA
# ============================================================================

print("\n" + "="*60)
print("📦 DATALOADER HAZIRLAMA")
print("="*60)

batch_size = 32

train_dataset = IrisDataset(dataset_path, split='train', processor=processor)
val_dataset = IrisDataset(dataset_path, split='val', processor=processor)
test_dataset = IrisDataset(dataset_path, split='test', processor=processor)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print(f"✅ Train batches: {len(train_loader)}")
print(f"✅ Val batches: {len(val_loader)}")
print(f"✅ Test batches: {len(test_loader)}")


# 6. EĞİTİM DÖNGÜSÜ
# ============================================================================

def train_epoch(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch in tqdm(dataloader, desc="🔄 Eğitim"):
        pixel_values = batch['pixel_values'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()

        outputs = model(pixel_values=pixel_values, labels=labels)
        loss = outputs.loss

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        predictions = outputs.logits.argmax(dim=-1)
        correct += (predictions == labels).sum().item()
        total += labels.size(0)

    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total

    return avg_loss, accuracy


def validate(model, dataloader, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="✅ Doğrulama"):
            pixel_values = batch['pixel_values'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(pixel_values=pixel_values, labels=labels)
            loss = outputs.loss

            total_loss += loss.item()

            predictions = outputs.logits.argmax(dim=-1)
            correct += (predictions == labels).sum().item()
            total += labels.size(0)

    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total

    return avg_loss, accuracy


# Optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)

# Eğitim
num_epochs = 10

print("\n" + "="*60)
print("🚀 EĞİTİM BAŞLIYOR")
print("="*60)

history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': []
}

best_val_acc = 0.0

for epoch in range(num_epochs):
    print(f"\n📍 Epoch {epoch+1}/{num_epochs}")
    print("-" * 60)

    train_loss, train_acc = train_epoch(model, train_loader, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, device)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(f"📊 Train Loss: {train_loss:.4f} | Train Acc: {train_acc*100:.2f}%")
    print(f"📊 Val Loss: {val_loss:.4f} | Val Acc: {val_acc*100:.2f}%")

    # En iyi modeli kaydet
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), '/content/best_vit_model.pth')
        print(f"💾 En iyi model kaydedildi! (Val Acc: {val_acc*100:.2f}%)")

print("\n" + "="*60)
print("✅ EĞİTİM TAMAMLANDI!")
print("="*60)


# 7. TEST VE CONFUSION MATRIX
# ============================================================================

print("\n" + "="*60)
print("🧪 TEST SETİ DEĞERLENDİRMESİ")
print("="*60)

# En iyi modeli yükle
model.load_state_dict(torch.load('/content/best_vit_model.pth'))
model.eval()

all_predictions = []
all_labels = []

with torch.no_grad():
    for batch in tqdm(test_loader, desc="🔍 Test ediliyor"):
        pixel_values = batch['pixel_values'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(pixel_values=pixel_values)
        predictions = outputs.logits.argmax(dim=-1)

        all_predictions.extend(predictions.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Metrikleri hesapla
test_accuracy = (np.array(all_predictions) == np.array(all_labels)).mean()
print(f"\n🎯 Test Accuracy: {test_accuracy*100:.2f}%\n")

# Classification report
print("📋 Detaylı Sınıflandırma Raporu:")
print("="*60)
print(classification_report(all_labels, all_predictions, target_names=class_names))

# Confusion Matrix
cm = confusion_matrix(all_labels, all_predictions)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names,
            cbar_kws={'label': 'Görüntü Sayısı'})
plt.title('Confusion Matrix - Test Seti\nVision Transformer (ViT)', fontsize=16, fontweight='bold')
plt.ylabel('Gerçek Sınıf', fontsize=12)
plt.xlabel('Tahmin Edilen Sınıf', fontsize=12)
plt.tight_layout()
plt.savefig('/content/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n💾 Confusion matrix kaydedildi: /content/confusion_matrix.png")


# 8. EĞİTİM GRAFİKLERİ
# ============================================================================

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss grafiği
axes[0].plot(history['train_loss'], label='Train Loss', marker='o')
axes[0].plot(history['val_loss'], label='Val Loss', marker='s')
axes[0].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Accuracy grafiği
axes[1].plot([acc*100 for acc in history['train_acc']], label='Train Accuracy', marker='o')
axes[1].plot([acc*100 for acc in history['val_acc']], label='Val Accuracy', marker='s')
axes[1].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy (%)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/training_history.png', dpi=300, bbox_inches='tight')
plt.show()

print("💾 Eğitim grafikleri kaydedildi: /content/training_history.png")


# 9. ÖZET
# ============================================================================

print("\n" + "="*60)
print("📊 PROJE ÖZETİ")
print("="*60)
print(f"✅ Model: {model_name}")
print(f"✅ Sınıflar: {', '.join(class_names)}")
print(f"✅ Train görüntü sayısı: {len(train_dataset)}")
print(f"✅ Val görüntü sayısı: {len(val_dataset)}")
print(f"✅ Test görüntü sayısı: {len(test_dataset)}")
print(f"✅ En iyi Val Accuracy: {best_val_acc*100:.2f}%")
print(f"✅ Test Accuracy: {test_accuracy*100:.2f}%")
print(f"✅ Kullanılan cihaz: {device}")
print(f"✅ Kaydedilen dosyalar:")
print(f"   - /content/best_vit_model.pth")
print(f"   - /content/confusion_matrix.png")
print(f"   - /content/training_history.png")
print("="*60)
```
